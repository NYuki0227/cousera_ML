```python
import numpy as np
import sklearn
import sklearn.decomposition as dec
import sklearn.datasets as ds
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = ds.load_iris()
X = iris.data
y = iris.target
print(X.shape)
```

```python
plt.scatter(X[:,0], X[:,1], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
X_bis = dec.PCA().fit_transform(X)
```

```python
print(X_bis.shape)
```

```python
plt.scatter(X_bis[:,0], X_bis[:,1], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
plt.scatter(X_bis[:,1], X_bis[:,2], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
plt.scatter(X_bis[:,2], X_bis[:,3], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
X_ter = dec.KernelPCA(kernel='rbf').fit_transform(X)
plt.scatter(X_ter[:,0], X_ter[:,1], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
plt.scatter(X_ter[:,1], X_ter[:,2], c = y, s = 30, cmap = plt.cm.rainbow)
```

```python
plt.scatter(X_ter[:,2], X_ter[:,3], c = y, s = 30, cmap = plt.cm.rainbow)
```
